```python
outliers_count = {}

for column in df.columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers_count[column] = ((df[column] < lower_bound) | (df[column] > upper_bound)).sum()

outliers_count
```

```python
#FITUR-FITUR numerik

list_numerik_features = ['Pregnancies', 'Glucose', 'BloodPressure', 
                         'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
```

```python
# Membuat boxplot untuk melihat outlier setiap fitur numerik
plt.figure(figsize=(14, 8))
for i, feature in enumerate(list_numerik_features):
    plt.subplot(4, 2, i + 1)
    sns.boxplot(x=df[feature], color='skyblue')
    plt.title(f'Boxplot of {feature}')
    plt.xlabel(feature)

plt.tight_layout()
plt.show()
```

```python
# handling outliers
def handle_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Menentukan batas bawah dan batas atas
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Mengganti nilai outlier dengan nilai batas atas atau batas bawah terdekat
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])

# Menggunakan fungsi handle_outliers_iqr untuk setiap kolom yang memiliki outlier
list_numerik_features
for column in list_numerik_features:
    handle_outliers_iqr(df, column)
```

```python
# Membuat boxplot untuk melihat outlier setiap fitur numerik
plt.figure(figsize=(14, 8))
for i, feature in enumerate(list_numerik_features):
    plt.subplot(4, 2, i + 1)
    sns.boxplot(x=df[feature], color='skyblue')
    plt.title(f'Boxplot of {feature}')
    plt.xlabel(feature)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.utils import resample, shuffle

# Pisahkan data berdasarkan nilai 'Outcome'
zero = df[df['Outcome'] == 0]
one = df[df['Outcome'] == 1]

# Upsampling kelas minoritas (Outcome = 1)
df_minority_upsampled = resample(one, 
                                 replace=True,     # sampel dengan penggantian
                                 n_samples=len(zero), # sesuaikan jumlah kelas mayoritas
                                 random_state=101) # untuk mendapatkan hasil yang sama setiap kali

# Gabungkan kembali kelas mayoritas dan hasil upsampling dari kelas minoritas
df_balanced = pd.concat([zero, df_minority_upsampled])

# Acak urutan data
df_balanced = shuffle(df_balanced, random_state=101) 

# Memeriksa distribusi kelas setelah penyeimbangan
print(df_balanced['Outcome'].value_counts())
```
